```python
#Encontrar la cantidad de veces que se repite la cadena de texto en un texto
def count_substring(string, sub_string):
    #contador
    contador = 0
    len_string = len(string)
    len_sub_string = len(sub_string)

    # resultado = all(item in lista_grande for item in lista_pequena)

    if len_string >= 1 and len_sub_string <= 200:
        if len_string >= len_sub_string:
            for i in range(0, len_string):
                if string[i:i+len(sub_string)] == sub_string:
                    contador+= 1
    return contador

if __name__ == '__main__':
    string = input().strip()
    sub_string = input().strip()

    count = count_substring(string, sub_string)
    print(count)
```

```python
# String Validator - verifica si una cadena de texto, cuenta con 5 formatos

if __name__ == '__main__':
    s = input()

    is_alnum = any([i.isalnum() for i in s])
    is_alpham = any([i.isalpha() for i in s])
    is_digit = any([i.isdigit() for i in s])
    is_lower = any([i.islower() for i in s])
    is_upper = any([i.isupper() for i in s])

    print(is_alnum)
    print(is_alpham)
    print(is_digit)
    print(is_lower)
    print(is_upper)

```
